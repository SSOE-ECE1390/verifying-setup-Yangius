```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = "Data/Apollo_11_Launch.jpg"
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.title("Original (Grayscale)")
plt.axis('off')
plt.show()
```

```python
h, w = img.shape
ch, cw = int(h * 0.6), int(w * 0.6)
y1, x1 = (h - ch) // 2, (w - cw) // 2
crop = img[y1:y1+ch, x1:x1+cw]

plt.imshow(crop, cmap='gray')
plt.title("Cropped (center 60%)")
plt.axis('off')
plt.show()
```

```python
brightness_delta = 100
brighter = cv2.add(img, np.full_like(img, brightness_delta))

plt.imshow(brighter, cmap='gray')
plt.title(f"Brightness +{brightness_delta}")
plt.axis('off')
plt.show()
```

```python
h, w = img.shape
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 30, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))

plt.imshow(rotated, cmap='gray')
plt.title("Rotated +30°")
plt.axis('off')
plt.show()
```
